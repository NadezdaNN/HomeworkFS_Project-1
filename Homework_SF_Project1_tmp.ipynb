# Преобразование данных

1. Создаем новый признак **"Образование"**, который будет иметь четыре категории: «высшее», «неоконченное высшее», «среднее специальное» и «среднее». Выясняем сколько соискателей имеет средний уровень образования.

```python
def education(x):    
    y=x.lower().split()
    out = y[0]+' '+y[1]
    out = out.replace('образование','')
    return out

data['Образование'] = data["Образование и ВУЗ"].apply(education)
#data['Образование'].unique()
print('Уровни образования:\n', data['Образование'].value_counts())
data=data.drop('Образование и ВУЗ', axis=1)
```

2. Создаем два новых признака **"Пол"** и **"Возраст"**. Определяем объем женских резюме (в процентах) и средний возраст соискателей.

```python
data['Пол'] = data["Пол, возраст"].apply(lambda x: x[0])
data['Возраст'] = data["Пол, возраст"].apply(lambda x: int(x.split(',')[1].split()[0]))
print(round(data['Пол'].value_counts(normalize=True)*100,2))
print('\nСредний возраст соискателей: ', round(data['Возраст'].mean(),1))
data=data.drop('Пол, возраст', axis=1)
```

3. Создаем новый признак  **"Опыт работы (месяц)"**, который будет содержать информацию о том, сколько месяцев проработал соискатель.

```python
def work(x):      
    if x is np.nan or x == 'Не указано':        
        return None    
    out=0
    out1=0
    y=x.split(' ')[:7]
    #y=y[:7]   
    for i,word in enumerate(y): 
        if word in ['год','года','лет']:                          
            out=int(y[i-1])             
        if word=='месяц' or word=='месяца' or word=='месяцев':                            
            out1=int(y[i-1])                         
    return int(out*12+out1)

data['Опыт работы (месяц)'] = data["Опыт работы"].apply(work)
print('Медианный опыт работы соискателей (в месяцах) равен ', round(data["Опыт работы (месяц)"].median()))
data=data.drop('Опыт работы', axis=1)
```

4. Создаем отдельные признаки **«Город»**, **«Готовность к переезду»**, **«Готовность к командировкам»**.

```python
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]

def sity(x):
    y=x.split(' ')
    if y[0] == 'Москва' or y[0] =='Санкт-Петербург':
        return y[0]
    if y[0] in million_cities:
        y[0]='город-миллионник'
    else:
        y[0]='другие'
    return y[0]

def removal(x):    
    tmp=['не готов к переезду', 'не готова к переезду']         
    if tmp[0] in x or tmp[1] in x:
        return False    
    else:
        return True

def businessTrip(x): 
    tmp=['не готов к командировкам', 'не готова к командировкам']    
    if tmp[0] in x or tmp[1] in x: 
        return False
    else:
        if 'командировка' in x:
            return True
        else:   # когда информация о командировках "потерялась"
            return False

data['Город'] = data["Город, переезд, командировки"].apply(sity)
data['Готовность к переезду'] = data["Город, переезд, командировки"].apply(removal)
data['Готовность к командировкам'] = data["Город, переезд, командировки"].apply(businessTrip)
#print(data['Готовность к командировкам'].value_counts)

print('Процентное соотношение соискателей по городам:\n', round(data['Город'].value_counts(normalize=True)*100))
#print(data['Готовность к переезду'].value_counts(normalize=True))
#print(data['Готовность к командировкам'].value_counts())
mask1=data['Готовность к переезду']==True
mask2=data['Готовность к командировкам']==True
print('\nПроцент соискателей, готовых и к переезду, и к командировкам: ', round(data[mask1 & mask2].shape[0]/data.shape[0]*100))

data=data.drop("Город, переезд, командировки", axis=1)
```

5. Создание признаков-«мигалок» для каждой категории столбцов **"Занятость"** и **"График"**.

```python
data['полная занятость'] = data["Занятость"].apply(lambda x: True if 'полная занятость' in x else False)
data['частичная занятость'] = data["Занятость"].apply(lambda x: True if 'частичная занятость' in x else False)
data['проектная работа'] = data["Занятость"].apply(lambda x: True if 'проектная работа' in x else False)
data['стажировка'] = data["Занятость"].apply(lambda x: True if 'стажировка' in x else False)
data['волонтерство'] = data["Занятость"].apply(lambda x: True if 'волонтерство' in x else False)

mask1=data['проектная работа']==True
mask2=data['волонтерство']==True
print("Количество людей, ищущих проектную работу и волонтёрство: ", data[mask1 & mask2].shape[0])

#print(data["График"].value_counts())
data['полный день'] = data["График"].apply(lambda x: True if 'полный день' in x else False)
data['гибкий график'] = data["График"].apply(lambda x: True if 'гибкий график' in x else False)
data['сменный график'] = data["График"].apply(lambda x: True if 'сменный график' in x else False)
data['удаленная работа'] = data["График"].apply(lambda x: True if 'удаленная работа' in x else False)
data['вахтовый метод'] = data["График"].apply(lambda x: True if 'вахтовый метод' in x else False)
print('\nКоличество людей, которые хотят работать вахтовым методом и с гибким графиком: ', data[(data['вахтовый метод']==True) & (data['гибкий график']==True)].shape[0])

data=data.drop("Занятость", axis=1)
data=data.drop("График", axis=1)
```

6.  Создание нового признака **«ЗП (руб)»** — заработная плата в рублях

```python
data_curr = pd.read_csv('../data/ExchangeRates.csv',sep=',')

data['Обновление резюме'] = pd.to_datetime(data['Обновление резюме']).dt.date
data_curr['date'] = pd.to_datetime(data_curr['date']).dt.date
#display(data)
#display(data_curr)

def currency(x):
    list1=['грн', 'USD', 'EUR', 'белруб', 'KGS', 'сум', 'AZN', 'KZT', 'руб']
    list2=['UAH', 'USD', 'EUR', 'BYN', 'KGS', 'UZS', 'AZN', 'KZT', 'RU']
    y=x.split()[1].replace('.', '')    
    for i,elem in enumerate(list1):
        if (elem in y):
            return list2[i]           

data['currency'] = data['ЗП'].apply(currency)
data['ЗП'] = data['ЗП'].apply(lambda x: float(x.split()[0]))
#display(data.head())
#print(data['currency'].value_counts(), data['currency'].count())

data_merge = data.merge(
    data_curr,   
    left_on = ['currency', 'Обновление резюме'], 
    right_on = ['currency', 'date'],    
    how='left'
)
#print(data_merge['close'].value_counts(), data_merge['close'].count())
#print(data_merge['proportion'].value_counts(), data_merge['proportion'].count())

data['ЗП (руб)']=data_merge['ЗП']*(data_merge['close'].fillna(1))/(data_merge['proportion'].fillna(1))
print('Желаемая медианная заработная плата соискателей (в рублях)', round(data['ЗП (руб)'].median()/1000), 'тыс.')

data=data.drop("ЗП", axis=1)
data=data.drop("currency", axis=1)

#data.info()
```

6. Строим распределение признака **"Возраст"** в логарифмическом масштабе. Добавляем к графику линии, отображающие среднее и границы интервала метода трех сигм (делаем послабление на 1 сигму в правую сторону).

```python
fig, axes = plt.subplots(1,1,figsize=(8,4))
log_age = np.log(data['Возраст']+1)
histplot = sns.histplot(log_age, bins=40, ax=axes)
histplot.axvline(log_age.mean(), color='k', lw=2)
histplot.axvline(log_age.mean()+4*log_age.std(), color='k', ls='--', lw=2)
histplot.axvline(log_age.mean()-3*log_age.std(), color='k', ls='--', lw=2)
histplot.set_title('распределение признака "Возраст" в логарифмическом масштабе');
print(log_age.skew()) # выводим численный показатель асимметрии
```

Распределение признака "Возраст" имеет правостороннюю асимметрию, т.к. её численный показатель больше 0.

```python
def outliers_z_score(data, feature, left=3, right=3, log_scale=False):
    if log_scale:
        x=np.log(data[feature]+1)
    else:
        x=data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left*sigma
    upper_bound = mu + right*sigma
    outliers = data[(x<lower_bound) | (x>upper_bound)]
    cleaned = data[(x>=lower_bound) & (x<=upper_bound)]
    return outliers, cleaned

outliers, cleaned_data = outliers_z_score(data, "Возраст", left=3, right=4, log_scale=True)
print('Число выбросов по методу z-отклонения: ', outliers.shape[0])
display(outliers)
```

Соискатели возраста 15 и 100 лет попали в категорию выбросов. Однако, резюме соискателя на должность "тестировщик игр" заслуживает более пристального внимая и может не являться "выбросом".
